```python
from tokenizer import WordpieceTokenizer
from torch.utils.data import Dataset, DataLoader
from dataset import TedDataset
import numpy as np
# Load data
with open('./dataset/de-en/train.en',encoding='utf-8') as f:
    en = f.readlines()
with open('./dataset/de-en/train.de',encoding='utf-8') as f:
    de = f.readlines()
    
# Load tokenizer
de_tokenizer = WordpieceTokenizer('de').load_model()
en_tokenizer = WordpieceTokenizer('en').load_model()

length_list = np.zeros(1000,dtype=int)
```

```python
for line in de:
    x = de_tokenizer.transform(line)
    length_list[len(x)] +=1
```

```python
for line in en:
    x = en_tokenizer.transform(line)
    length_list[len(x)] +=1
```

```python
lengths = []
counts = []
for length, count in enumerate(length_list):
    lengths.append(length)
    counts.append(count)
    print(length, count)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(lengths,counts)
plt.xlabel('length')
plt.ylabel('count')
plt.title('disease mention length statistic')
plt.show()
```
